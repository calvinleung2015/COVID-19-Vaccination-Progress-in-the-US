# Data Cleaning

```python
indo = df[df['country'] == 'United States']
indo.isna().sum()
```

```python
indo.head()
```

```python
grouped = indo[['date', 'people_vaccinated_per_hundred', \
              'people_fully_vaccinated', 'people_vaccinated', \
              'people_fully_vaccinated_per_hundred', \
              'daily_vaccinations_raw']]

grouped = grouped.set_index('date')

#inplace: replace NaN to 0.0
grouped.loc[:, 'people_fully_vaccinated'].fillna(0.0, inplace = True)
grouped.loc[:, 'people_fully_vaccinated_per_hundred'].fillna(0.0, inplace = True)
grouped.loc[:, 'daily_vaccinations_raw'].fillna(0.0, inplace = True)

dailies = {}
dailies['United States'] = grouped
```

```python
indo_df = pd.DataFrame(dailies['United States'])
indo_df
```

```python
indo_df.dropna(axis=0, inplace = True)
indo_df.iloc[:, 1:3] = indo_df.iloc[:, 1:3] / 1000
indo_df.iloc[:, 4] = indo_df.iloc[:, 4] / 1000
```

```python
#rename the variables
columns = {'people_vaccinated_per_hundred': 'people vaccinated per hundred',
           'people_fully_vaccinated': 'people fully vaccinated',
           'people_vaccinated': 'people vaccinated',
           'people_fully_vaccinated_per_hundred': 'people fully vaccinated per hundred',
           'daily_vaccinations_raw': 'daily vaccinations raw'}

indo_df.rename(columns = columns, inplace = True)
```

# Data Visualization & Analysis

```python
style.use('seaborn-bright')
plt.rc('xtick', labelsize=18) 
plt.rc('ytick', labelsize=18)
plt.rcParams['figure.figsize'] = (15, 8)
indo_df[['people vaccinated', 'people fully vaccinated']].plot(linewidth = 3)
plt.xlabel('date')
plt.ylabel('(in 1000) vaccinated')
plt.grid()
```
